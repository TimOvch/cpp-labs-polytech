```python
import matplotlib.pyplot as plt
from collections import defaultdict

def hash_function(word, table_size):
    hash_value = 0
    prime = 31
    for char in word:
        hash_value = (hash_value * prime + ord(char)) % table_size
    return hash_value

def visualize_collisions(words, table_size):
    table = defaultdict(list)
    collision_counts = [0] * table_size

    for word in words:
        index = hash_function(word, table_size)
        table[index].append(word)
        collision_counts[index] = len(table[index]) - 1  # Увеличиваем количество коллизий

    # Визуализация
    plt.bar(range(table_size), collision_counts)
    plt.title('Распределение коллизий в хеш-таблице')
    plt.xlabel('Индекс в таблице')
    plt.ylabel('Количество коллизий')
    plt.show()

# Большой список слов
words = [
    "ананас", "мандарин", "яблоко", "вишня", "груша", 
    "пиня", "папайя", "малина", "помидор", "персик",
    "апельсин", "вишня", "клубника", "абрикос", "киви", 
    "манго", "грейпфрут", "банан", "гранат", "персик",
    "дуриан", "финик", "слива", "кокос", "черешня", 
    "лайм", "лимон", "клюква", "черника", "облепиха",
    "черная смородина", "морошка", "голубика", "ягода", 
    "морковка", "кабачок", "баклажан", "редис", "свекла", 
    "петрушка", "укроп", "кресс-салат", "шпинат", "помидор", 
    "репа", "тихоходка", "очень", "крупа", "молоко", 
    "сок", "чай", "кофе", "мясо", "рыба", 
    "салат", "сахар", "соль", "перец", "клубника", 
    "вишня", "ежевика", "крыжовник", "персик", "нектарин",
    "груша", "апельсин", "мандарин", "лайм", "грейпфрут", 
    "брокколи", "петрушка", "шуруп", "фортепиано", "собака",
    "кошка", "птица", "рыба", "слон", "тигр", 
    "лев", "жираф", "медведь", "заяц", "волк",
    "кролик", "черепаха", "носорог", "антилопа", "козел",
    "коровы", "овцы", "птицы", "кенгуру", "мартышка", 
    "слон", "акула", "кит", "медуза", "дельфин",
    "белуга", "мандарин", "кабан", "лиса", "сорока",
    "помело", "фейхоа", "имбирь", "кмина", "чеснок", 
    "петрушка", "укроп", "тайская", "базилик", "орегано",
    "масло", "вода", "соль", "перец", "орех",
    "фундук", "миндаль", "аромат", "вкус", "процесс",
    "суть", "пункт", "важность", "необходимость", "событие"
]

table_size = 20

# Визуализация коллизий
visualize_collisions(words, table_size)
```
