```python
import numpy as np
import matplotlib.pyplot as plt

def get_random_cauchy(mu, lambda_param, size=10000):
    """Генерирует случайные числа из распределения Коши."""
    return mu + lambda_param * np.tan(2 * np.pi * np.random.rand(size))

# Параметры распределения
mu = 5
lambda_param = 1.1

# Генерация данных
data = get_random_cauchy(mu, lambda_param, size=1000)
data = data[(data >= -60) & (data <= 60)]

# Построение гистограммы
plt.figure(figsize=(10, 6))
plt.hist(data, bins=50, density=True, alpha=0.6, color='g')

# Настройка графика
plt.title(f'Гистограмма распределения Коши\nmu = {mu}, lambda = {lambda_param}')
plt.xlabel('Значение')
plt.ylabel('Плотность')
plt.axvline(mu, color='r', linestyle='dashed', linewidth=2, label=f'mu = {mu}')
plt.text(mu, 0.02, f'mu = {mu}', color='red', fontsize=12, ha='center')


# Сохранение гистограммы как файла
plt.grid()
plt.savefig('cauchy_distribution_histogram.png')
plt.show()
```
